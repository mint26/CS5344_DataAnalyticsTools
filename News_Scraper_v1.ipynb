## Query by Text Search
## Below code will scrape for 500k tweets between 22 July, 2022 and 23 August, 2022, by a text search ("$BBBY") then save results in a csv file

```python
# Setting variables to be used below
maxTweets = 500000

# Creating list to append tweet data to
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in tqdm(enumerate(sntwitter.TwitterSearchScraper('$BBBY since:2022-07-22 until:2022-08-23').get_items())):
    if i > maxTweets:
        break
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
```

```python
# Creating a dataframe from the tweets list
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])

# Display dataframe
tweets_df
```

```python
####final update by John on dynamic web link according to interested stocks########
# using selenium for web scraping on articles links from CNBC search
options = Options()
options.add_argument('--window-size=1920,1080')
options.add_argument('--headless')
options.add_argument(f'--user-agent={ua}')#,'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8', 'Accept-Language': 'en-US,en;q=0.5', 'Accept-Encoding': 'gzip, deflate', 'DNT': '1', 'Connection': 'close', 'Upgrade-Insecure-Requests': '1'")
#s = Service(r'C:\Users\65911\AppData\Local\Temp\Temp1_chromedriver_win32.zip\chromedriver.exe')
#s = r'C:\Users\65911\AppData\Local\Temp\Temp1_chromedriver_win32.zip\chromedriver.exe'
s = Service('C:/Users/John Wong/chromedriver.exe')
driver = webdriver.Chrome(service=s, options=options)
#driver = webdriver.Chrome(executable_path=s, options=options)

####For dynamic link#############
#note that there were 2 google stocks initially but condensed to one here. also included for meta and facebook due to name change.
top25_list = ['Apple', 'Microsoft', 'Amazon', 'Tesla', 'Google', 'Nvidia', 'Berkshire Hathaway', 'Meta', 'Facebook', 'UnitedHealth', 'Johnson & Johnson', 'Jpmorgan', 'Visa', 'Procter & Gamble', 'Exxon Mobil',
              'Home Depot', 'Chevron Corporation', 'Mastercard', 'Bank of america', 'Abbvie', 'Pfizer', 'Broadcom', 'Costco Wholesale', 'Disney', 'Coca-Cola']

url_list = []
temp_url_list = []
for i in range(25):
    company = top25_list[i]
    company_href = company.replace(" ","%20").replace("&","%26")
    myUrl = 'https://www.cnbc.com/search/?query=' + company_href + '%20earnings%20preview&qsearchterm=' + company_href + '%20earnings%20preview'
    #example of 2 words link:
    #https://www.cnbc.com/search/?query=apple%20orange%20earnings%20preview&qsearchterm=apple%20orange%20earnings%20preview  
    #print(myUrl)
    driver.get(myUrl)
    sleep(5)  # this site requires minimum 5 seconds wait to load finish before attempting to get number of search results and scroll down
    ####End For dynamic link#############

    ######get exact number of articles to scrape based on number of search results######
    num_articles = driver.find_element(by=By.XPATH, value='//*[@id="searchresultsamount"]').text
    num_articles = int(num_articles.split()[0])
    print(f'number of search results: {num_articles}')  # for testing reference only
    print(type(num_articles))  # for testing reference only

    ######to deal with inifinity loop######
    scroll_pause_time = 1  # You can set your own pause time.
    screen_height = driver.execute_script("return window.screen.height;")  # get the screen height of the web
    scroll_count = 1
    #num_scroll_down = 5

    while True:
      #if scroll_count == num_scroll_down:
          #break
        # print current scroll count for tracking of progress in while loop
        if scroll_count%10 == 0:
            print(f'current scroll count in while loop: {scroll_count}')
        # scroll one screen height each time
        driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=scroll_count))  
        scroll_count += 1
        sleep(scroll_pause_time)
        # update scroll height each time after scrolled, as the scroll height can change after we scrolled the page
        scroll_height = driver.execute_script("return document.body.scrollHeight;")  
        # break the loop when the height we need to scroll to is larger than the total scroll height
        if (screen_height) * scroll_count > scroll_height:
            break
        #scroll_count += 1
        #####end#######

    for div in tqdm(range(num_articles)):  # updated to num_articles based on exact number of search results

        #####retry up to 3 times if fail to get url######
        no_element_count = 0
        for j in range(3):
            try:
                elem3 = driver.find_element(by=By.XPATH, value="//*[@id='searchcontainer']/div["+str(div+2)+"]/div/div[2]/div[2]/a")
                url = elem3.get_attribute('href')
                print(url)  # for reference
                print(elem3)  # for reference
                if url:
                    sleep(1)
                    break
            except NoSuchElementException:
                try:
                    elem3 = driver.find_element(by=By.XPATH, value="//*[@id='searchcontainer']/div["+str(div+2)+"]/div/div/div[2]/a")
                    url = elem3.get_attribute('href')
                    print(url)  # for reference
                    print(elem3)  # for reference
                    if url:
                        sleep(1)
                        break
                except NoSuchElementException:               
                    if no_element_count == 3:
                        break
                    no_element_count += 1
                    sleep(1)

        # if extracted url is duplicate in temp_url_list, continue loop without storing in url_list
        if temp_url_list and url in temp_url_list:
            continue

        # only store article url in url_list
        if url[:26] != 'https://www.cnbc.com/video' and url[:24] != 'https://buffett.cnbc.com' and url[:27] != 'https://www.cnbc.com/select':
            url_dict = {'Company': company,
                        'URL': url}
            
            # append each url_dict entry to url_list
            url_list.append(url_dict)
            # append each url to temp_url_list for comparison to store only unique url later
            temp_url_list.append(url)
            #sleep(3)
        #######end#########

driver.close()

url_list
```

```python
#####Final code updated by John#####
# create news_data function to get analyst articles texts from url website

def news_data(company, url):
    news_list = []
    article_list = []
    skip_box_count = 0
    full_text_joined = ''
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.71 Safari/537.36', 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8', 'Accept-Language': 'en-US,en;q=0.5', 'Accept-Encoding': 'gzip, deflate', 'DNT': '1', 'Connection': 'close', 'Upgrade-Insecure-Requests': '1'}
    page_request = requests.get(url, headers=headers)
    soup = BeautifulSoup(page_request.content, "html.parser")
    # check if article is PRO article, skip PRO article as full text is not available
    skip_url = soup.find('a', {'class': 'ProPill-proPillLink'})
    #print(f'skip_url: {skip_url}')
    # check if article is CLUB article, skip CLUB article as full text is not available
    skip_url2 = soup.find('a', {'class': 'InvestingClubPill-investingClubPillLink'})    
    #print(f'skip_url2: {skip_url2}')
    # check if article is not available, skip article as text is not available
    try:
        skip_url3 = soup.body.h1.text
    except AttributeError:
        skip_url3 = None
    #print(f'skip_url3: {skip_url3}')
    
    try:
        skip_url4 = soup.find('h1', {'class': 'PageHeader-title'}).text
    except AttributeError:
        skip_url4 = None        
    #print(f'skip_url4: {skip_url4}')
    
    try:
        skip_url5 = soup.find('h1', {'class': 'ArticleHeader-headline'}).text
    except AttributeError:
        skip_url5 = ''        
    #print(f'skip_url5: {skip_url5}')
    
    if skip_url == None and skip_url2 == None and skip_url3 != '503 Service not available' and skip_url4 != 'CNBC TV' and skip_url5[:20] != 'Coronavirus updates:':
        #print('test_text')
        # get article date in DD MMM YYYY format
        try:
            date_search = soup.find('time', {'itemprop': 'datePublished'})
            #print(f'date_search: {date_search}')
            date_text = date_search.text.strip()
            #print(f'date_text: {date_text}')
            comma_idx = date_text.find(',')
            #print(f'comma_idx: {comma_idx}')
            # check for cnbc logo label in url
            cnbc_logo = soup.find('span', {'id': 'sourcelogo-cnbccom'})
            #print(f'cnbc_logo: {cnbc_logo}')
            # check for cnbc contents label in url
            cnbc_content = soup.find('div', {'id': 'cnbc-contents'})
            #print(f'cnbc_content: {cnbc_content}')

            # check for comma index to find date_text position to slice date
            if comma_idx != -1:
                if date_text[comma_idx+1] == ' ' and date_text[comma_idx+2] != ' ':
                    #print('comma+1, not comma+2')
                    date_text_sliced = date_text[comma_idx+2:].strip()
                if date_text[comma_idx+1] == ' ' and date_text[comma_idx+2] == ' ':
                    #print('comma+1, comma+2')
                    date_text_sliced = date_text[comma_idx+3:].strip()
                if date_text[comma_idx+1] != ' ':
                    #print('not comma+1')
                    date_text_sliced = date_text[comma_idx+1:].strip()

                # if cnbc_logo or cnbc_content label found, set date as date_text_sliced
                if cnbc_logo != None or cnbc_content != None:
                    date = date_text_sliced                  
                #print(f'date_text_sliced: {date_text_sliced}')
                
                # if cnbc_logo and cnbc_content label not found, further slice date text to get date
                if cnbc_logo == None and cnbc_content == None: #Sep 6 201612:00 AM EDT Sep 16 2016
                    colon_idx = date_text_sliced.find(':')
                    #print(f'colon_idx: {colon_idx}')
                    # check for colon index to narrow down string position to slice date part
                    try:
                        if int(date_text_sliced[4])%int(date_text_sliced[4]) == 0 and int(date_text_sliced[5])%int(date_text_sliced[5]) == 0:
                            date_text_sliced = date_text_sliced[:11]
                            date = date_text_sliced[4:6] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]
                    except ZeroDivisionError:
                        if int(date_text_sliced[4]) != 0 and int(date_text_sliced[5]) == 0:
                            date_text_sliced = date_text_sliced[:11]
                            date = date_text_sliced[4:6] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]
                        if int(date_text_sliced[4]) == 0 and int(date_text_sliced[5]) != 0:
                            date_text_sliced = date_text_sliced[:11]
                            date = date_text_sliced[4:6] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]                            
                    except ValueError:
                        if date_text_sliced[3] == ' ':
                            if int(date_text_sliced[4])%int(date_text_sliced[4]) == 0:
                                date_text_sliced = date_text_sliced[:10]
                                date = date_text_sliced[4] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]
                        
            # if comma index not found, set date as date_text less 'published' word for further data processing later on
            else:
                date_text = date_text.lstrip('published').strip()
                if date_text[:3] == 'Mon' or 'Tue' or 'Wed' or 'Thu' or 'Fri' or 'Sat' or 'Sun':
                    date_text_sliced = date_text[5:]
                    try:
                        if int(date_text_sliced[4])%int(date_text_sliced[4]) == 0 and int(date_text_sliced[5])%int(date_text_sliced[5]) == 0:
                            date_text_sliced = date_text_sliced[:11]
                            date = date_text_sliced[4:6] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]
                    except ZeroDivisionError:
                        if int(date_text_sliced[4]) != 0 and int(date_text_sliced[5]) == 0:
                            date_text_sliced = date_text_sliced[:11]
                            date = date_text_sliced[4:6] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]
                        if int(date_text_sliced[4]) == 0 and int(date_text_sliced[5]) != 0:
                            date_text_sliced = date_text_sliced[:11]
                            date = date_text_sliced[4:6] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]
                    except ValueError:
                        if date_text_sliced[3] == ' ':
                            if int(date_text_sliced[4])%int(date_text_sliced[4]) == 0:
                                date_text_sliced = date_text_sliced[:10]
                                date = date_text_sliced[4] + ' ' + date_text_sliced[:3] + ' ' + date_text_sliced[-4:]
        
        # if no date found, set date as empty
        except AttributeError:
            #print('test_date')
            date = ''            
        if len(date) > 11 and date[2] == ' ':
            date_clean = date[:6] + ' ' + date[-4:]
        else:
            date_clean = date
        #print(f'date_clean: {date_clean}')
            
        div_group = soup.findAll('div', {'class': 'group'})
        #print(f'div_group: {div_group}')
        sleep(1) # can add in sleep here if there is HTTP error scraping data too quickly

        # loop through all the div groups in article
        for group in div_group:
            skip_box = group.find('div', {'id': 'RegularArticle-RelatedContent-1'})
            #print(f'skip_box: {skip_box}')
            # if skip box with irrelevant text is found, continue loop without storing text (need to skip group twice)
            if skip_box != None or skip_box_count == 1:
                if skip_box_count == 0:
                    skip_box_count = 1
                    continue
                if skip_box_count == 1:
                    skip_box_count = 0
                    continue                
            text_list = []
            # loop through each row in group to extract article text
            for row in group:
                if row:
                    text = row.text.strip()
                    text_list.append(text)

            # join text in text_list
            text_joined = ' '.join(text_list)
            text_joined = text_joined.strip()
            # append joined text in news_list
            news_list.append(text_joined)

        # join full text in news_list
        full_text_joined = ' '.join(news_list).strip()

        # if valid text is found, create article_dict to store url and article text
        article_dict = {'Company': company,
                        'URL': url,
                        'Articles': full_text_joined,
                        'Date': date_clean}
        # append article_dict in article_list to store each article text
        article_list.append(article_dict)
    
    # if invalid article or invalid text is found, do not append to article_list and articles_df later on       
    if skip_url or skip_url2 or skip_url3 == '503 Service not available' or skip_url4 == 'CNBC TV' or skip_url5[:20] == 'Coronavirus updates:' or full_text_joined[:53] == ' bringing the six-week total to more than 30 million.':
        #print('full_text_test')
        article_list = []
    
    return article_list
```
